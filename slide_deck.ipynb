> In my presentation I want to emphazise the fact that students with immigration background in the second-generation have a much less sense of belonging to school than the native students or students with first-generation immigration status when they perform less than the modal grades in the country. 

## Dataset Overview

> "PISA is a survey of students' skills and knowledge as they approach the end of compulsory education. It is not a conventional school test. Rather than examining how well students have learned the school curriculum, it looks at how well prepared they are for life beyond school.
Around 510,000 students in 65 economies took part in the PISA 2012 assessment of reading, mathematics and science representing about 28 million 15-year-olds globally. Of those economies, 44 took part in an assessment of creative problem solving and 18 in an assessment of financial literacy."

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
pisa = pd.read_csv('pisa2012.csv', encoding='latin-1')
```

```python
# drop columns of no interest
pisa = pisa[['GRADE', 'IMMIG', 'BELONG']]

# drop null values
pisa.dropna(inplace = True)

# get indices of rows that will be dropped and drop rows with these indices
drp_idx = pisa.index[pisa['GRADE'] == 3.0].tolist()
pisa = pisa.drop(drp_idx)
```

## Worldwide distribution of students with immigration backgrounds in schools

> The data shows that roughly 90% of the students are natives, 5% have an immigration status of 'First-Generation' and 5% of 'Second-Generation'.

```python
# set figure size
plt.figure(figsize = [10, 10])

# create pie chart for the distribution of students with immigration backgrounds in school
sorted_counts = pisa['IMMIG'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False);
plt.axis('square');

# add title to the pie chart
plt.title('Worldwide distribution of students with immigration backgrounds in schools');
```

## Average sense of belonging to the school by immigration status

> Here we can see that students that are immigrants in the first generation they state on average a less sense of belonging to the school they are in than students that are native or immigrants in the second generation.

```python
# set figure size
plt.figure(figsize = [10, 10])

# set base color
base_color = sb.color_palette()[0]

# set variable for displayed order of immigration groups
gen_order = ['First-Generation', 'Native', 'Second-Generation']


# plotting a barplot with the sense of belonging to school and the uncertainty whiskers for different immigration groups
sb.barplot(data = pisa, x = 'IMMIG', y = 'BELONG', color = base_color, order = gen_order)

# axis labeling
plt.xticks (rotation = 15);
plt.ylabel('Avg. sense of belonging to school');
plt.xlabel('Immigration status');

# add title to chart
plt.title('Average sense of belonging to the school by immigration status');
```

## Sense of belonging to school separated by grade and immigration status

> The data reveals that for the grade values of -1.0 and less the sense of belonging to school is on average much lower for the second-generation immigration students than for both the first-generation immigration and the native students. For the grades of 1.0 and more the sense of belonging to school is on average even higher for second-generation immigration students. It seems that beeing an immigrant in the second generation in a country the sense of belonging to school is strengthened in either direction, for worse grades to a lower sense of belonging to school and for better grades to a higher sense of belonging to school. 
> It is surprising that not the first-generation immigration students have a lower sense of belonging to school when performing bad, but the second-generation students. 

```python
# set figure size
plt.figure(figsize = [10, 10])

# plotting a multivariate barplot 
sb.barplot(data = pisa, x = 'GRADE', y = 'BELONG', hue = 'IMMIG');
plt.xlabel('Grade compared to modal grade in the country')
plt.ylabel('Avg. sense of belonging to school');

# add title to chart
plt.title('Avg. sense of belonging to school for different immigration groups and grades');
```
